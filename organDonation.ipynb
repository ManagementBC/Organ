```python

import openai
import time
import json
import requests
import concurrent.futures

# ✅ Initialize the Smart Contract
contract = w3.eth.contract(address=contract_address, abi=contract_abi)

# ✅ OpenAI API Setup
openai.api_key = "xxxxxxxxxxxxxxxxxxxxxxx"
MODEL_ID = "ft:gpt-4o-mini-2024-07-18:personal:organdonations:AmbvC8Vu"

# ✅ Pinata IPFS Gateway
PINATA_GATEWAY = "https://gateway.pinata.cloud/ipfs/"

# ✅ Fetch donor IPFS hash from smart contract
def get_donor_ipfs(donor_id):
    donor_id = int(donor_id)
    if donor_id < 1:
        return None
    donor_data = contract.functions.donors(donor_id).call()
    return donor_data[5] if donor_data[7] else None  # IPFS hash index (Ensuring ethical approval)

# ✅ Fetch all recipients and their IPFS hashes
def get_all_recipients():
    recipient_count = contract.functions.recipientCounter().call()
    print(f"🔍 Total Recipients: {recipient_count}")  # Debugging

    recipients = []
    for recipient_id in range(1, recipient_count + 1):
        recipient_data = contract.functions.recipients(recipient_id).call()

        # ✅ Debugging output
        print(f"👀 Recipient {recipient_id}: {recipient_data}")

        # ✅ Ensure recipient is registered, has an IPFS hash, and is ethically approved
        if recipient_data[6] and recipient_data[5] and recipient_data[8]:
            recipients.append({"recipient_id": recipient_data[0], "ipfs_hash": recipient_data[5]})

    print(f"✅ Final Recipients List: {recipients}")  # Debugging
    return recipients if recipients else {"error": "No recipients available"}


# ✅ Fetch data from IPFS (parallel processing for efficiency)
def fetch_from_ipfs(ipfs_hash):
    try:
        response = requests.get(PINATA_GATEWAY + ipfs_hash, timeout=5)
        return response.json() if response.status_code == 200 else None
    except requests.exceptions.RequestException:
        return None

def fetch_multiple_from_ipfs(ipfs_hashes):
    with concurrent.futures.ThreadPoolExecutor() as executor:
        results = list(executor.map(fetch_from_ipfs, ipfs_hashes))
    return [r for r in results if r]  # Remove None values

# ✅ Match donor with 2 best recipients using GPT-4o-mini
def match_donor(donor_id):
    try:
        start_time = time.time()  # Start timing

        donor_ipfs = get_donor_ipfs(donor_id)
        if not donor_ipfs:
            return {"error": "Donor data not found on IPFS"}

        donor_data = fetch_from_ipfs(donor_ipfs)
        if not donor_data:
            return {"error": "Failed to retrieve donor data from IPFS"}

        recipient_list = get_all_recipients()
        if not recipient_list:
            return {"error": "No recipients available"}

        recipient_data = fetch_multiple_from_ipfs([r["ipfs_hash"] for r in recipient_list])
        if not recipient_data:
            return {"error": "No valid recipient data found"}

        # ✅ Prepare GPT-4 prompt
        prompt = {"donor": donor_data, "recipients": recipient_data}

        response = openai.ChatCompletion.create(
            model=MODEL_ID,
            messages=[
                {
                    "role": "system",
                    "content": "You are an AI designed to match organ donors to recipients. Return the output strictly in JSON format without extra text. The output format must be: {'donor_id': 1, 'top_matches': [{'recipient_id': 2}, {'recipient_id': 3}]}",
                },
                {"role": "user", "content": json.dumps(prompt)},
            ],
        )

        # ✅ Debugging: Print response from OpenAI
        print("🔍 OpenAI Response:", response)

        # ✅ Validate response
        if not response or "choices" not in response or not response["choices"]:
            return {"error": "Invalid OpenAI response"}

        response_content = response["choices"][0]["message"]["content"].strip()
        if not response_content:
            return {"error": "OpenAI response is empty"}

        try:
            match_result = json.loads(response_content.replace("'", '"'))
        except json.JSONDecodeError:
            return {"error": "Failed to parse OpenAI response. Response might be empty or not JSON formatted."}

        donor_id = match_result["donor_id"]
        recipient_id1 = match_result["top_matches"][0]["recipient_id"]
        recipient_id2 = match_result["top_matches"][1]["recipient_id"]

        # ✅ Record match in smart contract
        tx_hash = contract.functions.createMatch(donor_id, recipient_id1, recipient_id2).transact({
            "from": w3.eth.default_account,
            "gas": contract.functions.createMatch(donor_id, recipient_id1, recipient_id2).estimateGas({"from": w3.eth.default_account}),
        })

        w3.eth.wait_for_transaction_receipt(tx_hash)

        elapsed_time = time.time() - start_time  # Calculate execution time
        print(f"✅ Matching completed in {elapsed_time:.2f} seconds")

        return match_result

    except json.JSONDecodeError:
        return {"error": "Failed to parse OpenAI response. Response might be empty or not JSON formatted."}
    except openai.error.OpenAIError as e:
        return {"error": f"OpenAI API error: {str(e)}"}
    except Exception as e:
        return {"error": f"Unexpected error: {str(e)}"}

# ✅ Gradio UI with Debug Mode & Title
interface = gr.Interface(
    fn=match_donor,
    inputs=[gr.Textbox(label="Enter Donor ID")],
    outputs=[gr.JSON(label="Matching Results")],
    title="Organ Matching System",
    description="Enter a donor ID to find the best two recipient matches based on compatibility, urgency, and location.",
)

interface.launch(debug=True)


```
